## Basic plotting with matplotlib

````python
x = np.linspace(0, 5. * np.pi, 100)
y = np.sin(x)
````

````python
plt.figure()
plt.plot(x, y)
plt.plot(x, y + 1)
````

What is happening here:

    plt.figure(): Create a new figure
    plt.plot(...): Add a plot to this figure
    Magic command `%matplotlib inline` makes plots being shown at the end of a cell's execution automatically
    
Why use `plt.figure`?

If you don't create a new figure explicitly, then plots are added to the existing figure.


**Tips**: Use trailing ";" to remove the output line [<matplotlib.lines.Line2D at 0x19871617320>]

````python
plt.plot(x, y)
plt.plot(x, y + 1);
````

### Line/Marker styles in Matplotlib

There are 2 ways to set the style for the line and the points on the edges

1. Specify each property as its own argument in the plot function

    ```python
    plt.plot(x, y,
            linestyle='dashed', linewidth=10, color='blue',
            marker='o', markersize=5, markerfacecolor='green', markeredgecolor='red'
            )
    ```

2. Use an abbreviation ([documented here](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot))

    ```python
    plt.plot(x, y, 'bo--')  # dash line, 'o' marker, blue
    ```

matlab like command-line style plotting, line plot

````python
plt.plot(x, y, 'b--', label='line1')
plt.plot(x, y + 1, 'r-', label='line2');
# add x,y axis labels, title and legends
plt.xlabel("The X-axis")
plt.ylabel("The Y-axis")
plt.title('A title')
plt.legend()
````

Line plot with markers only

````python
plt.plot(x, y, marker='o', markersize=6, label='line1')
plt.plot(x, y + 1, marker='v', markersize=6, label='line2');
# add x,y axis labels, title and legends
plt.xlabel("The X-axis")
plt.ylabel("The Y-axis")
plt.title('A title')
plt.legend()
````

### Pyplot vs Objects

There are two ways to interact with Matplotlib plots

When we use `plt.plot` to plot a picture, we're using the pyplot state machine.

This was developed to give a MATLAB-like interface to the plotting system in matplotlib.

`plt.plot` creates a plot using the **current axes** in the **current figure**.

We could also call methods on these Figures and Axes directly.

### Figures and Axes objetcs

The second way of plotting is to get access to the figures and axes handle and then add plots to each axes

1. Get the figures and axes handle by calling plt.subplots()
2. Calls ax.plot(...) to add the plot to the axes `ax`.


````python
fig, ax = plt.subplots()
ax.plot(x, y, 'b--', label='line1')
ax.plot(x, y + 1, 'r', label='line2');
ax.legend()
ax.set(title='A title', xlabel='The X-axis', ylabel='The Y-axis')
````

### Adjusting subplots

You'll notice that the plots aren't positioned very nicely.

There are three ways to fix this

1. Give the plots more space to start with
    - When you call `plt.figure`, you can specify a figure size <br><br>
    
2.  Manually specify different spacings using `plt.subplots_adjust`
    - left = 0.125
        - the left side of the subplots of the figure
    - right = 0.9
        - the right side of the subplots of the figure
    - bottom = 0.1
        - the bottom of the subplots of the figure
    - top = 0.9
        - the top of the subplots of the figure
    - wspace = 0.2
        - the amount of width reserved for blank space between subplots, expressed as a fraction of the average axis width
    - hspace = 0.2
        - the amount of height reserved for white space between subplots, expressed as a fraction of the average axis height<br><br>
3.  Call `plt.tight_layout()` and let matplotlib figure it out

````python
fig, axes = plt.subplots(figsize=[10,6], nrows=2, ncols=2)
ax = list(axes.ravel())
idx = 0
ax[idx].plot(x, y, 'b--',label='line1')
ax[idx].plot(x, y + 1,'r', label='line2');
ax[idx].legend(bbox_to_anchor=(1, 1))
ax[idx].set(title='A title', xlabel='The X-axis', ylabel='The Y-axis');

idx = 1
ax[idx].plot(x-1, y, 'b--',label='line1')
ax[idx].plot(x, y + 1,'r', label='line2');
ax[idx].legend(bbox_to_anchor=(1, 1))
ax[idx].set(title='A title', xlabel='The X-axis', ylabel='The Y-axis');

idx = 2
ax[idx].plot(x+1, y, 'b--',label='line1')
ax[idx].plot(x, y + 1,'r', label='line2');
ax[idx].legend(bbox_to_anchor=(1, 1))
ax[idx].set(title='A title', xlabel='The X-axis', ylabel='The Y-axis');

idx = 3
ax[idx].plot(x, y, 'b--',label='line1')
ax[idx].plot(x, y - 1,'r', label='line2');
ax[idx].legend(bbox_to_anchor=(1, 1))
ax[idx].set(title='A title', xlabel='The X-axis', ylabel='The Y-axis');

# Make it look prettier!!
plt.tight_layout()
````

### Gridspec for more complicated figure layouts

So far we just showed subplots using multiple axes that evenly divided the plot area.

If you wanted to make a more complicated layout you can use GridSpec

````python
from matplotlib import gridspec
plt.figure(figsize=[12,8])

# Complicated layouts
gs = gridspec.GridSpec(3, 3)

ax1 = plt.subplot(gs[0, :])   # Use row 1, all columns
ax2 = plt.subplot(gs[1,:-1])  # Use row 2, all columns but the last one
ax3 = plt.subplot(gs[1:, -1]) # Use row 2&3, only the last column
ax4 = plt.subplot(gs[-1,0])   # Use the last row, first column
ax5 = plt.subplot(gs[-1,-2])  # Use the last row, second column

plt.subplots_adjust(wspace=.3, hspace=.3)
````

### Scatter plots

Difference between scatter and plot
- Use scatter plots when you want every point to have a different size or color
- Let's see an example

````python
tips = sns.load_dataset('tips')
tips.head()
````

````python
sns.set_context("poster", font_scale=1.2)
plt.scatter(tips.total_bill, tips.tip, s=30)
plt.xlabel("total_bill")
plt.ylabel("tip")
````

Using color to represent a 3rd variable 

````python
plt.scatter(x='total_bill', y='tip',data=tips, c='size',cmap='RdBu_r', s=30)
plt.xlabel("Total_bill")
plt.ylabel("Tip")
plt.colorbar()
````

### Histograms

Lets look at some histograms in matplotlib

````python
plt.figure()
plt.hist(tips.total_bill, bins=20);
plt.xlabel("Total_bill")
plt.ylabel("Counts")
````

Add multiple histograms - use 'alpha' to overlay

````python
plt.figure()
plt.hist(tips.total_bill, bins=20, label='Total_bill', alpha=.6)
plt.hist(tips.tip, bins=20, label='Tip', alpha=.6)
plt.xlabel("Amount")
plt.ylabel("Counts")
plt.legend()
````

````python
tips_agg = tips.groupby(['sex', 'time'])['total_bill','tip'].sum().rename(columns={'tip':'Tip_sum'})
tips_agg.head()
````

````python
tips_agg.plot.pie(y='Tip_sum',figsize=(6,6),autopct='%.2f%%')
plt.legend([])
````

````python
tips_agg2.plot.pie(y='Tip_mean',figsize=(6, 6),autopct='%.2f%%')
plt.legend([])
````

Observation: But on average, waitors earned as much as waitresses

## Seaborn

Seaborn is a plotting library that is built on top of **matplotlib**

You can do anything in seaborn with just matplotlib commands.  Seaborn just makes it much less tedious.

Seaborn is useful for:

- Heatmaps
- Statistical plots
- Dealing with Categorical variables

````python
import seaborn as sns  # This is how seaborn is usually abbreviated
sns.set_context("poster", font_scale=1.2)
sns.set_style("darkgrid")
````

````python
tips = sns.load_dataset('tips')
tips.head()
````

### Seaborn *jointplot*

`jointplot` is a handy function for plotting the joint distribution of two variables

````python
sns.jointplot(tips.total_bill, tips.tip, size=8);
````

Here's our plot! Some things to notice:

- Seaborn has already filled in the x and y labels
- This is actually a matplotlib Figure.  The figure has three Axes (subplots)

````python
sns.jointplot(x=tips.total_bill, y=tips.tip, kind='kde',size=8);
````

### DataFrames and Seaborn

Seaborn makes it easy to just use dataframes directly.

Recall that our 'tips' dataframe has columns 'total_bill' and 'tip'

All seaborn plotting functions provide an alternate way of calling them that uses the dataframe directly:

````python
# Use dataframe columns directly
sns.jointplot(x='total_bill', y='tip', data=tips, size=8)
````

### Seaborn *regplot/lmplot* 
Visualizing linear relationships

````python
sns.regplot(x="total_bill", y="tip", data=tips);
````

lmplot conditioning on 1 categorical variable

````python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
````

lmplot conditioning on 2 categorical variables

````python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
````

lmplot conditioning on 3 categorical variables

````python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips);
````

### Seaborn *pairplot*

`pairplot` is a handy function for plotting multiple pairwise distributions in a dataset

````python
sns.pairplot(vars=['total_bill', 'tip', 'size'], data=tips,size=3);
````

Plot pairwise distribution and condition on 1 categorical variable

````python
sns.pairplot(vars=['total_bill', 'tip', 'size'], hue='sex', data=tips)
````
